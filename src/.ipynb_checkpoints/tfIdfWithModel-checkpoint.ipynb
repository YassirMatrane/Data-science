```python
import numpy as np
from sklearn.utils import shuffle
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.svm import LinearSVC
from sklearn.naive_bayes import MultinomialNB,GaussianNB
from sklearn.metrics import accuracy_score, recall_score,f1_score,confusion_matrix,classification_report
from sklearn.metrics import precision_recall_fscore_support as score
from matplotlib import pyplot as plt
import seaborn as sns
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import SGDClassifier
from sklearn.ensemble import BaggingClassifier
import pickle
```

```python
df = pd.read_csv("../dataSet/trainingDataFinal.csv")
df = shuffle(df)


df['class'] = df['hate'] + df['offensive']*2
```

```python
n=len(df)
#6000 for testing data
m=6000
y_train = list(df["class"].values[:n-m])
y_train_test = list(df["class"].values[n-m:len(df)])

x_train = df["comment"].values[:n-m]
x_val = df["comment"].values[n-m:len(df)]

# Calculate TF-IDF
tf_idf = TfidfVectorizer(ngram_range=(1, 5),
                         binary=False,sublinear_tf=True)
X_train_tfidf = tf_idf.fit_transform(x_train)
X_val_tfidf = tf_idf.transform(x_val)
with open('../vectorizers/vectorizerHateOffense.pk', 'wb') as fin:
    pickle.dump(tf_idf, fin)
```

# tfIdf With naive bayes

```python
Naive = MultinomialNB()
Naive.fit(X_train_tfidf,y_train)
# predict the labels on validation dataset
predictions_NB = Naive.predict(X_val_tfidf)
pickle.dump(Naive, open("../models/NaiBayOHeModel", 'wb'))
```

```python
# Use accuracy_score function to get the accuracy
print("Naive Bayes Accuracy Score -> ",accuracy_score(predictions_NB, y_train_test)*100)
print("Naive Bayes recall Score -> ",recall_score(predictions_NB, y_train_test,average='macro')*100)
print("Naive Bayes f1Measure Score -> ",f1_score(predictions_NB, y_train_test,average='macro')*100)
```

```python
cf_matrix = confusion_matrix(y_train_test, predictions_NB) 
fig, ax = plt.subplots(figsize=(15,10)) 
ax = sns.heatmap(cf_matrix, linewidths=1, annot=True, ax=ax, fmt='g')
ax = ax.set(ylabel="True Label", xlabel="Predicted Label", xticklabels=['0','1' ,'2', '3'], yticklabels=['0','1', '2', '3'])
```

```python
#prediction
comment = "اخرص يا أعرابي يا وقح فلن تعدو قدرك يا سافل "
comment2 = "يا مقرف يا جبان للأسف هذه تسمى خسة من شخص أحمق"

X_pred_tfidf = tf_idf.transform([comment2])
print(X_pred_tfidf)
print(type(X_pred_tfidf))
predictions_NB = Naive.predict(X_pred_tfidf)
predictions_NB
```

# tfIdf with random forest

```python
rfClassifier = RandomForestClassifier(n_estimators=70)
rfClassifier.fit(X_train_tfidf,y_train)
# predict the labels on validation dataset
predictions_NB = rfClassifier.predict(X_val_tfidf)
# Use accuracy_score function to get the accuracy
print("random Forest Accuracy Score -> ",accuracy_score(predictions_NB, y_train_test)*100)
print("random Fores recall Score -> ",recall_score(predictions_NB, y_train_test,average='micro')*100)
print("random Fores f1Measure Score -> ",f1_score(predictions_NB, y_train_test,average='micro')*100)
```

```python
from sklearn.metrics import confusion_matrix 
from matplotlib import pyplot as plt
import seaborn as sns
cf_matrix = confusion_matrix(y_train_test, predictions_NB) 
fig, ax = plt.subplots(figsize=(15,10)) 
ax = sns.heatmap(cf_matrix, linewidths=1, annot=True, ax=ax, fmt='g')
ax = ax.set(ylabel="True Label", xlabel="Predicted Label", xticklabels=['0','1' ,'2', '3'], yticklabels=['0','1', '2', '3'])
```

```python
import xgboost as xgb
classifier = xgb.XGBClassifier(
    n_estimators=80,
    reg_lambda=1,
    gamma=0,
    max_depth=25
)
classifier.fit(X_train_tfidf,y_train)
```

```python
# predict the labels on validation dataset
predictions_NB = classifier.predict(X_val_tfidf)
# Use accuracy_score function to get the accuracy
print("XGB Accuracy Score -> ",accuracy_score(predictions_NB, y_train_test)*100)
print("XGB recall Score -> ",recall_score(predictions_NB, y_train_test,average='micro')*100)
print("XGB f1Measure Score -> ",f1_score(predictions_NB, y_train_test,average='macro')*100)
```

```python
svm = LinearSVC()
svm.fit(X_train_tfidf,y_train)
# predict the labels on validation dataset
predictions_NB = svm.predict(X_val_tfidf)
# Use accuracy_score function to get the accuracy
print("Bagging decision Tree Accuracy Score -> ",accuracy_score(predictions_NB, y_train_test)*100)
print("Bagging decision Tree recall Score -> ",recall_score(predictions_NB, y_train_test,average='micro')*100)
print("Bagging decision Tree f1Measure Score -> ",f1_score(predictions_NB, y_train_test,average='micro')*100)
```

# Making two models for offensive and hate classes in order to solve the confusion problem between the two classes
