```python
def wordEm(text):
    print(text, end='...')
    text_preprocessed = preprocess(text,
                                    do_farasa_tokenization = True,
                                    farasa = farasa_segmenter,
                                    use_farasapy = True)
    arabert_input = arabert_tokenizer.encode(text_preprocessed,add_special_tokens=True)
    tensor_input_ids = torch.tensor(arabert_input).unsqueeze(0)
    output = arabert_model(tensor_input_ids)
    print('Converted')
    return output[0][0][1:-1]
```

Our model will have one input layer, one embedding layer followed by one LSTM layer with 100 neurons. The output from the LSTM layer will be used as the input to the 2 dense output layers. Each output layer will have 1 neuron with sigmoid activation function. Each output will predict integer value between 1 and 0 for the corresponding label.

The sigmoid activation function will return a value between 0 and 1 for each neuron. If any neuron's output value is greater than 0.5, it is assumed that the comment belongs to the class represented by that particular neuron.

```python
input_layer = Input(shape=(None, 768))
lstm_1 = Bidirectional(LSTM(100, return_sequences=False))(input_layer)
dense_1 = Dense(64, activation='relu')(lstm_1)
hate_output = Dense(1, activation='sigmoid')(dense_1)
offensive_output = Dense(1, activation='sigmoid')(dense_1)

model = Model(inputs=input_layer, outputs=[hate_output, offensive_output])
model.compile(optimizer='Adam', loss='binary_crossentropy')
```

In the second approach we will create one dense output layer for each label. Our model will have one input layer, one bididirectional layer, one LSTM layer with 64 neurons and one output layer with 2 neurons since we have 2 labels in the output.

```python
input_layer_2 = Input(shape=(None, 768))
lstm_1_2 = Bidirectional(LSTM(100, return_sequences=False))(input_layer_2)
dense_1_2 = Dense(64, activation='relu')(lstm_1_2)
dense_layer_1_2 = Dense(2, activation='sigmoid')(LSTM_Layer_1_2)
model_2 = Model(inputs=input_layer_2, outputs=dense_layer_1_2)

model_2.compile(loss='binary_crossentropy', optimizer='adam', metrics=['acc'])
```
